```python
# importing the required modules
import pandas as pd
import matplotlib.pyplot as plt
# specifying the path to csv files
path = r"C:\Users\Classes\Desktop\exp 6 data william kogen\6.2.3 BH relationship in the air\6.2.3 BH relationship in the air_01.csv"

def mv_corr(df):
    return (df / 1000)


np = 400
Lp = 0.042
Rp = 2.05
Ri = 9888
C = 450.6 * 10**-9
ns = 500
As = 4.16 * 10**-5



df = pd.read_csv(path)

Vy = df['Channel A'][1:].astype(float)
Vy = mv_corr(Vy)
Vx = df['Channel B'][1:].astype(float)

H = Vx * (np)/(Lp * Rp)
H = H.values.tolist()
B = Vy * ((Ri * C)/(ns * As))
B = B.values.tolist()

plt.plot(H, B)
plt.xlabel('H / Am^-1')
plt.ylabel('B / Teslas')
plt.title('B vs H for air')
plt.grid()
plt.show()

```
